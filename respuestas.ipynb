### 1.Importacion de dataset con pandas

```python
import pandas as pd

df = pd.read_csv("ventas.csv")
```

### 2. Exploracion Incial de los datos

```python
primeras_filas = df.head()
ultimas_filas = df.tail()

print("Primeras:")
print(primeras_filas)

print("Ultimas:")
print(ultimas_filas)
```

### 3. Limpieza y normalizacion de los datos

3.1 eliminar filas con valores nulos

```python
df = df.dropna()
df.head()
```

3.2 columna fecha pasarla a tipo datatime

```python
df["Fecha"] = pd.to_datetime(df["Fecha"],errors="coerce" )
df.info()
```

3.3 Corregir las mayusculas en el caso de Nombre y Apellodo

```python
df["Nombre_Cliente"] = df["Nombre_Cliente"].str.title()
df["Apellido_Cliente"] = df["Apellido_Cliente"].str.title()
df.head()
```

### 4. Obtener de Estadisticas Descriptivas

4.1 Calculas las estadisticas descriptivas

```python
df.describe()
```

4.2 Calculas estadisticas por grupo

```python
df.groupby("ID_Venta").describe()
```

### 5. Visualizacion de los Datos con Matplotlib

```python
import matplotlib.pyplot as plt
plt.hist(df["Apellido_Cliente"], bins=20, color="skyblue")
plt.title("Apellidos de los clientes")
```
